```python
# import pyMongo
from pymongo import MongoClient

#connect to mongodb; creating instance of MongoClient
client = MongoClient()

#create or access database
db = client.video

#access collection movieDetails
md = db.movieDetails

print(md.find_one({"director":'George Lucas'}))

#serverStatusResult = db.command("serverStatue")
#print(serverStatusResult)
```

```python
# If we want to read all the rows then we use cursor , like this:

for i in md.find({"director":'George Lucas'}): # this is cursor
    print(i)
```

```python
### Exercise : display the title and year of all movies released in year 2000

for i in md.find({'year':2000},{'title':1,'year':1,'_id':0}):
    print("The Movie",i['title'],"is released on year",int(i['year']))
```

```python
# list only 8th and 9th document

for i in md.find({'year':2000},{'title':1,'year':1,'_id':0}).limit(2).skip(7):
    print("The Movie",i['title'],"is released on year",int(i['year']))
```

```python
# name of movies in which Dennis Hopper is acted

for i in md.find({'actors':'Dennis Hopper'},{'title':1,'_id':0}):
    print("The Movie are",i['title'])
```

```python
# give title and actors of the movies releasing after 1990

for i in md.find({'year':{'$gt':1990}},{'title':1,'actors':1,'_id':0}):
    print("The Movie are",i['title'],'and actors are',i['actors'])
```

```python
# give title and actors of the movies releasing after 1990 and sorted in based on year

for i in md.find({'year':{'$gt':1990}},{'title':1,'actors':1,'_id':0}).sort('year',1):
    print("The Movie are",i['title'],'and actors are',i['actors'])
```

```python
# # give title of the movies betweeen 1990 and 2000

for i in md.find({'year':{'$gte':1990,'$lte':2000}},{'title':1,'year':1,'_id':0}).sort('year',1):
    print("The Movie name is",i['title'],'and it released on',int(i['year']))
```

```python
# for every movies display their IMDB rating


for i in md.find({},{'tomato.rating':1,'title':1,'_id':0}):
    print(i)
```

```python
# for every movies display their IMDB rating greater than 9


for i in md.find({'imdb.rating':{'$gt':9}},{'imdb.rating':1,'title':1,'_id':0}):
    print("The Movie name is",i['title'],'and its rating is',int(i['imdb']['rating']))
```

```python
# write a query in movies if the attribute year exists then listed

for i in md.find({'year':{'$exists':True}},{'year':1,'_id':0}).sort('year'):
    print(i)
```

```python
# write a query in movies if the attribute year exists then listed

for i in md.find({'year':{'$exists':True}},{'year':1,'_id':0}).sort('year').distinct('year'):
    print(i)
```

```python
# OR

for i in md.distinct('year'):
    print(i)
```

```python
# in movie detail, list the movie either imdb rating >=9 or they were not release in year 2000

for i in md.find({'$or':[{'imdb.rating':{'$gte':9}},{'year':{'$ne':2000}}]},{'_id':0,'title':1,'year':1}).sort('year',1):
    print(i['title'])
```

```python
# for every year wise what is the maximum nomination received

for i in md.aggregate([{'$group':{'_id':'$year','nomination':{'$max':'$awards.nominations'}}},{'$sort':{'_id':-1}}]):
    print(i)
```

```python
# for every year wise what is the count of movies

for i in md.aggregate([{'$group':{'_id':'$year','movies':{'$sum':1}}},{'$sort':{'_id':-1}}]):
    print(i)
```

```python
# for every year wise which has the runtime >=100 and count them

for i in md.aggregate([{'$match':{'runtime':{'$gte':100}}},{'$group':{'_id':'$year','count':{'$sum':1}}},{'$sort':{'_id':1}}]):
    print(i)
```

### Cities Json

```python
# import pyMongo
from pymongo import MongoClient

#connect to mongodb; creating instance of MongoClient
client = MongoClient()

#create or access database
db = client.USA

#access collection movieDetails
ct = db.city

print(ct.find_one({}))

#serverStatusResult = db.command("serverStatue")
#print(serverStatusResult)
```

```python
# 6. List the cities and their ranks. Sort it based on their city names.

for i in ct.find({},{'_id':0,'city':1,'rank':1}).sort('city',1):
    print('The rank of the City',i['city'],'is',i['rank'])
```

```python
# 7.For every state, list the city having highest latitude.

for i in ct.aggregate([{'$group':{'_id':'$state','Highest_Latitude':{'$max':'$latitude'}}}]):
    print('The Highest Latitude of the city',i['_id'],'is',i['Highest_Latitude'])
```

```python
# 8. Write a python program to display the states uniquely.

print('The Unique states are:')
for i in ct.distinct('state'):
    print(i)
```

```python
# 9. Write a python program to display number of cities in each state.

for i in ct.aggregate([{'$group':{'_id':{'state':'$state'},'count':{'$sum':1}}},{'$sort':{'_id':1}}]):
    print('The Number of Cities in the',i['_id']['state'],'is',i['count'])
```

```python
# 10. Write a python program to list the cities along with their latitude and longitude that have latitude>=30 and <=50 and 
# longitude>-71.

for i in ct.find({'$and':[{'latitude':{'$gte':30,'$lte':50}},{'longitude':{'$gt':-71}}]
                 },{'_id':0,'city':1,'latitude':1,'longitude':1}):
    print(i)
```
